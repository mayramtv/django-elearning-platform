```python
from faker import Faker
from faker.providers import BaseProvider
from faker.providers import DynamicProvider
import pandas as pd
import random
```

# Student

```python
# costumize provider
email_domain = DynamicProvider(provider_name='email_domain',
                               elements = ['gmail', 'yahoo', 'example', 'university', 'hotmail', 'mail'])

email_tl_domain = DynamicProvider(provider_name='email_ext',
                                  elements=['com', 'edu', 'org', 'online'])
# initiate faker instance
fake = Faker()
# add provider to faker 
fake.add_provider(email_domain)
fake.add_provider(email_tl_domain)
```

```python
data_list = []
for i in range(10):
    f_name = fake.first_name()
    l_name = fake.last_name()
    u_name = (f_name + l_name).lower() + str(random.randrange(0, 9))
    birtday = fake.date_between(start_date='-40y', end_date='-15y')
    data_dict = {
        'account_number': 'AC' + f_name[0] + l_name[0] + birtday.strftime('%m%d%y'),
        'name': f_name,
        'last_name': l_name,
        'birthday': birtday,  
        'email': u_name + '@' + fake.email_domain() + '.' + fake.email_ext(),
        'user_name': u_name,
        'start_date': fake.date_between(start_date='-5y', end_date='-1y'),
        'photo':  u_name + '_pic.jpg',
    } 
    data_list.append(data_dict)

data_profile = pd.DataFrame(data_list)
data_profile.to_csv('Data/Students.csv', index=False)
```

# Teachers

# Courses

```python
# create provider for techers choice
teachers_list = list(data_teacher_profile['account_number'])
teacher_acc = DynamicProvider(provider_name='teacher_acc',
                                  elements=teachers_list)
# initiate faker instance
fake = Faker()
# add provider to faker 
fake.add_provider(teacher_acc)

c_names = {
    'computer science': ['python', 'javascript', 'web development'],
    'english language': ['writting', 'english', 'essay writting'],
    'business management': ['business analytics', 'statistics'],
}
courses = {}
codes = []
names = []
category = []
teachers = []
created_dates = []
updated_dates = []
descriptions = []
imgs = []
# Create table for courses
for i, key in enumerate(c_names.keys()):
    for j, name in enumerate(c_names[key]):
        for k, val in enumerate(['introduction to ', 'intermediate ', 'advance ']):
            codes.append(key.split(' ')[0][0] + key.split(' ')[1][0] + str (i + 1) + '0' + str(j) + str(k + 1))
            names.append(val + name)
            category.append(key)
            teachers.append(fake.teacher_acc())
            startdate = fake.date_between(start_date='-8y', end_date='-1y')
            created_dates.append(startdate)
            updated_dates.append(fake.date_between(startdate, 'today'))
            descriptions.append('This course is for ' + val + name + ' in ' + key)
            imgs.append(name + str(k) + '_img.jpg')
# add lists to dictionary
courses['code'] = codes            
courses['name'] = names
courses['category'] = category
courses['teacher'] = teachers
courses['date_created'] = created_dates
courses['date_updated'] = updated_dates
courses['description'] = descriptions
courses['image'] = imgs
courses_df = pd.DataFrame(courses) 
courses_df.to_csv('Data/Courses.csv', index=False)
print(len(courses_df))
courses_df.head()
```

## Courses' images 

```python
from PIL import Image
from PIL import ImageDraw
from PIL import ImageFont 

hues = ['blue', 'green', 'orange']

for i, cat in enumerate(courses_df['category']):
    pos = (40, 40)
    
    if cat == 'computer science':
        col = fake.color(hue=hues[0], color_format='rgb', luminosity='light')
    if cat == 'english language':
        col = fake.color(hue=hues[1], color_format='rgb', luminosity='light')
    if cat == 'business management':
        col = fake.color(hue=hues[2], color_format='rgb', luminosity='light')
        pos = (30, 40)
        
    text = courses_df['name'][i].replace(' ', '\n')
    font1 = ImageFont.truetype("Oswald-VariableFont_wght.ttf", 15)
    font2 = ImageFont.truetype("Oswald-VariableFont_wght.ttf", 20)

    img = Image.new(mode="RGB", size=(200, 200), color=col)
    imgDraw = ImageDraw.Draw(img)
    # Add text
    imgDraw.multiline_text(pos, courses_df['category'][i].title(), fill=(250, 250, 250), align="center", font=font2)
    imgDraw.multiline_text((50, 80), text.title(), fill=(0, 0, 0), align="center", font=font2)
    # img.show()
    img.save("Data/courses_images/" + courses_df['image'][i])
    
```

# Resources

```python
resources_dict = {}
resourcesTypes_dict = {}
couseIDs = []
codes = []
times = []
resourceTypes = []
contents = []
texts = []
for i, cde in enumerate(courses_df['code']):
    # create providers for faker
    dts = [courses_df['date_created'][i], courses_df['date_updated'][i]]
    date_created = DynamicProvider(provider_name='date_created', elements=dts)
    r_type = DynamicProvider(provider_name='resource_type', elements=['text', 'image', 'video'])
    imgs = DynamicProvider(provider_name='img_type', elements=['png', 'jpg']) 
    # add provider to model
    fake.add_provider(date_created)
    fake.add_provider(r_type)
    fake.add_provider(imgs)
    
    #iterate over the number of courses to create each reosurce
    for j in range(random.randrange(0, 3)):
        couseIDs.append(cde)
        
        codes.append(cde + '.' + str(float(j) + 1))
        # Create the time stamp
        if j == 0:
            times.append(courses_df['date_created'][i])
        else:
            times.append(fake.date_created())
        # resource type
        rtype = fake.resource_type()
        resourceTypes.append(rtype)
        # create resource name
        texts.append(fake.text())
        if rtype == 'image':
            contents.append(fake.file_name(category=rtype, extension='jpg'))
        elif rtype == 'text':
            contents.append(fake.file_name(category=rtype, extension='txt'))
        else:
            contents.append(fake.file_name(category=rtype, extension='mp4'))

resources_dict['code'] = codes
resources_dict['couse_ID'] = couseIDs
resources_dict['timestamp'] = times
resources_dict['content'] = contents
resources_dict['text'] = texts 
resources_df = pd.DataFrame(resources_dict) 
resources_df.to_csv('Data/Resources.csv', index=False)

# ResourcesTypes table
resourcesTypes_dict['couse_ID'] = couseIDs
resourcesTypes_dict['resource_type'] = resourceTypes
resourcesTypes_df = pd.DataFrame(resourcesTypes_dict) 
resourcesTypes_df.to_csv('Data/ResourcesTypes.csv', index=False)
```

```python
import cv2
import glob
path = resources_df['content']
for i, cat in enumerate(resourcesTypes_df['resource_type']):
    c_path = "Data/resources_files/" + path[i]

    if cat == 'text':
        # Open the file 
        with open(c_path, 'w') as file:
            # Add content
            file.write("Text file sample")
    if cat == 'image':
        img = Image.new(mode="RGB", size=(200, 200), color=fake.color(hue='blue'))
        imgDraw = ImageDraw.Draw(img)
        img.save(c_path)
    
    if cat == 'video':
        # Initialize video writer
        video = cv2.VideoWriter(c_path, cv2.VideoWriter_fourcc(*'mp4v'), 24, (200, 200))

        # Read images and write them to the video
        for file in sorted(glob.glob('images/*.jpg')):
            img = cv2.imread(file)
            # attaches images to video
            video.write(img)
        video.release()
        
```

```python
# create provider for course choice
course_list = list(courses_df['code'])
course_code = DynamicProvider(provider_name='course_code',
                                  elements=course_list)
# initiate faker instance
fake = Faker()
# add provider to faker 
fake.add_provider(course_code)

students = data_profile['account_number']
starts = data_profile['start_date']

enrollments = {}
timeStamps = []
courses_l =[]
# create a table with enrollment timestamp, student and course
for i, student in enumerate(students):
    timeStamps.append(fake.date_between(starts[i], 'today'))
    courses_l.append(fake.course_code())
enrollments['date'] = timeStamps
enrollments['student'] = students
enrollments['course'] = courses_l
enrollments_df = pd.DataFrame(enrollments)

enrollments_df.to_csv('Data/Enrollments.csv', index=False)    
enrollments_df
```
